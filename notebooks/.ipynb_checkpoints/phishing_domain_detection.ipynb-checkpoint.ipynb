### Importing Necessary Libraries

### Loading Dataset into a DataFrame

```python
data = "C://Users//SANTHOSH//Desktop//iNeuron//Phishing Domain Detection//data//dataset_full.csv"
df = pd.read_csv(data)
```

Removes all rows with missing (NaN) values from the DataFrame `df`.



```python
df = df.dropna()
df.head()
```

### Splitting Features and Target Variable

```python
y = df["phishing"]
x = df.drop("phishing", axis=1)
```

### Dropping Low-Correlation Features

```python
correlation_matrix = df.corr()


correlation_with_target = correlation_matrix['phishing']  

# Identify features with correlation less than 0.05
low_corr_features = correlation_with_target[correlation_with_target.abs() < 0.05].index

# Drop these features from the DataFrame
a = df.drop(columns=low_corr_features)

print(f"Dropped features: {low_corr_features}")
print(f"Remaining features: {a.columns}")
```

### Removing Features with Insignificant Variance

```python
from sklearn.feature_selection import VarianceThreshold

var = VarianceThreshold()
var.fit(x)

print("Total Columns :",len(var.get_support()))
print("Columns Dropped :",len(var.get_support()) - sum(var.get_support()))

X_new = x[x.columns[var.get_support()]]

print("Columns Left :",len(X_new.columns))
```

### Removing Less Correlated Features

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                col_corr.add(corr_matrix.columns[i])
    return col_corr
corr_features = correlation(x, .6)
sum(var.get_support()) - len(corr_features)
X_new = x.drop(corr_features, axis=1)
print("Number of Column Dropped", len(x.columns) - len(X_new.columns) )

X = X_new
print("Columns Left :",len(X_new.columns))
```

### Creating a Directory to Save the Model

### Saving Feature Names Using Joblib

```python
feature_names_path = os.path.join(save_dir, "Feature_Names.pkl")
joblib.dump(x.columns, feature_names_path)
print(f"Feature names saved to: {feature_names_path}")
```

### Splitting Data and Training the Model

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = HistGradientBoostingClassifier()
model.fit(X_train, y_train)
```

### Evaluating the Model Performance

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.4f}")
print(classification_report(y_test, y_pred))
```
